```python
# Reemplazo de valores negativos usando numpy
f = "../data/esolmet/esolmet.xlsx"
mes = "Abril"
def importar_mes(f, mes):
    
    df = pd.read_excel(f, sheet_name=mes, skiprows=2, index_col=0, parse_dates=True)
    df.index = pd.to_datetime(df.index, errors='coerce')
        
    return df

df_abril = importar_mes(f, mes)

# Función para reemplazar negativos

def reemplaza_negativos(df_abril):
    df_reemplazado = df_abril.copy()
    columnas_radiacion = ['Directa', 'Global', 'Difusa']
    for col in columnas_radiacion:
        # Convertir la columna a numérica (si no lo es), ignorando errores
        df_reemplazado[col] = pd.to_numeric(df_reemplazado[col], errors='coerce')
        # Reemplazar valores negativos por np.nan
        df_reemplazado.loc[df_reemplazado[col] < 0, col] = np.nan
    return df_reemplazado

df_reemplazado = reemplaza_negativos(df_abril)
df_reemplazado
```

```python
f = "../data/esolmet/esolmet.xlsx"
mes = "Abril"
def importar_mes(f, mes):
    
    df = pd.read_excel(f, sheet_name=mes, skiprows=2, index_col=0, parse_dates=True)
    df.index = pd.to_datetime(df.index, errors='coerce')
        
    return df

df = importar_mes(f, mes)

# función para renombrar columnas
def renombra_columnas(df):
    df_renombrado = df.copy()
    mapeo = {
        'Directa': 'Ib',
        'Global': 'Ig',
        'Difusa': 'Id',
        'Temperatura': 'To',
        'Humedad': 'HR',
        'Viento Vel.': 'VV',
        'Presiòn Bar.': 'P'
    }
    df_renombrado = df_renombrado.rename(columns=mapeo)
    # Reordenar
    df_renombrado.columns = [col for col in df_renombrado.columns]
    nuevo_orden = ['Ig', 'Id', 'Ib', 'UV', 'To', 'P', 'HR', 'VV']
    df_renombrado = df_renombrado[nuevo_orden]
    return df_renombrado

df_nuevo = renombra_columnas(df)
df_nuevo.head()


```

```python
def filtrar_columnas_con_I(df):
    columnas_filtradas = [col for col in df.columns if 'I' in col]
    return df[columnas_filtradas]

df_filtrado = filtrar_columnas_con_I(df_nuevo)
df_filtrado
```

```python
# Escribe una función llamada calcula_estadisticas que calcule y devuelva las estadísticas descriptivas básicas de una columna específica en un DataFrame de pandas. Esta función facilitará el acceso rápido a valores estadísticos clave para análisis adicionales.

def calcula_estadisticas(df, columna):
    # Convertir la columna a tipo numérico, ignorando errores
    datos = pd.to_numeric(df[columna], errors='coerce')
    # Calcular las estadísticas
    promedio = datos.mean()
    maximo = datos.max()
    minimo = datos.min()
    desviacion_estandar = datos.std()

    return {
        "promedio": promedio,
        "maximo": maximo,
        "minimo": minimo,
        "desviacion_estandar": desviacion_estandar
    }


info_estadisticas = calcula_estadisticas(df_nuevo, 'Ig')
info_estadisticas
```
