```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Figuras y ejes 
# fig, ax = plt.subplots()

file_path = "../../data/Cuernavaca_Enero_comas.csv"
df = pd.read_csv(file_path, index_col=0, parse_dates=True)

df.info()
```

```python
plt.plot(df.To)
```

```python
# La línea fig, ax = plt.subplots() crea una nueva figura (fig) y un conjunto de ejes (ax).
# Esto es útil para personalizar gráficos, ya que ax proporciona métodos para graficar datos y ajustar los ejes.
# En este caso, ax.plot(df.P) grafica la columna 'P' del DataFrame en los ejes creados.

fig, ax = plt.subplots()
ax.plot(df.P)
```
