```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '../../data/Cuernavaca_Enero_comas.csv'
df = pd.read_csv(file_path, index_col=0, parse_dates=True)
df.head()

```

```python
fig, ax = plt.subplots()

ax.plot (df.To, df.RH)
```

```python
fig, ax = plt.subplots()

ax.plot ("To", "RH", data=df)
```

```python
fig, ax = plt.subplots()
ax.scatter ("To", "RH", data=df)
```

```python
file_path = "../../data/Cuernavaca_1dia_comas.csv"
df1dia = pd.read_csv(file_path, index_col=0, parse_dates=True)
# Renombrar columnas
df1dia.columns = ['To', 'Ws', 'Wd', 'P', 'Ig', 'Ib', 'Id']
df1dia.head()
```
