```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "../../data/Cuernavaca_Enero_comas.csv"
df = pd.read_csv(file_path, index_col=0, parse_dates=True)
df.head()
```

```python
plt.plot(df.To)
plt.title("Temperatura en Cuernavaca - Enero 2023")
```

```python
# Gráfica de dispersión
# Compara como se comportan dos variables
# En este caso, Temperatura vs Humedad Relativa
plt.scatter(df.To, df.RH)
plt.title("Temperatura vs Humedad Relativa")
```
