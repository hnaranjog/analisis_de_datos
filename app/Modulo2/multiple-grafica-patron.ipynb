```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from dateutil.parser import parse
```

```python
file_path = "../../data/Cuernavaca_Enero_comas.csv"
dfEnero = pd.read_csv(file_path, index_col=0, parse_dates=True)
dfEnero.index
```

```python
dfEnero.columns
```

```python




fig = plt.figure(figsize=(12, 7))
gs = gridspec.GridSpec(2, 2)

ax1 = fig.add_subplot(gs[0,:])
ax1.set_title("Temperatura")

ax2 = fig.add_subplot(gs[1,:1])
ax2.set_title("Humedad Relativa")


ax3 = fig.add_subplot(gs[1,1:])
ax3.set_title("Radiacin")

ax1.plot(dfEnero.To)
ax2.scatter(dfEnero.To, dfEnero.RH)
ax3.plot(dfEnero.Ig)
ax3.plot(dfEnero.Ib)
ax3.plot(dfEnero.Id)

f1 = parse("2012-01-01")
f2 = f1 + pd.Timedelta("1d")
ax3.set_xlim(f1, f2)
```
