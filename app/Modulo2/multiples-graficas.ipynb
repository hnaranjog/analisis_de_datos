```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
file_path = "../../data/Cuernavaca_Enero_comas.csv"
dfEnero = pd.read_csv(file_path, index_col=0, parse_dates=True)
dfEnero.index
```

```python
dfEnero.head()
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(10, 3), sharex=True)
f1 = parse("2012-01-01")
f2 = f1 + pd.Timedelta("3d12h")

ax1.plot(dfEnero.Ig, label="Ig")
ax1.plot(dfEnero.Ib, label="Ib")
ax1.plot(dfEnero.Id, label="Id")
ax1.set_title("Radiación Solar en Cuernavaca")
ax1.legend()
ax1.grid()


ax2.plot(dfEnero.To, label="To")
ax2.set_title("Temperatura y Humedad en Cuernavaca")
ax2.legend()
ax2.grid()

ax2.set_xlim(f1, f2)

ax1.set_ylabel("Radiación($[W/m²]$)")
ax2.set_ylabel("T($[°C]$)")
ax2.set_xlabel("Tiempo")
```
