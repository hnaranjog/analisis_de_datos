```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
file_path = "../../data/Cuernavaca_Enero_comas.csv"
df1 = pd.read_csv(file_path, index_col=0, parse_dates=True)
df1.index
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

f1 = parse("2012-01-01")
f2 = f1 + pd.Timedelta("1d")

ax.plot(df1.Ig,"r-", label="Ig")
ax.plot(df1.Ib,"k--", label="Ib")
ax.plot(df1.Id,"bo-", label="Id")

ax.set_xlim(f1, f2)
ax.grid()
ax.legend()
ax.set_xlabel("Tiempo [mm-dd HH]")
ax.set_ylabel("Radiaci√≥n $[¬µW/m¬≤]$")
ax.set_title("Radiaci√≥n Solar en Cuernavaca")
```
