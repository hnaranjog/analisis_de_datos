```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from dateutil.parser import parse
```

```python
file_path = "../../data/Cuernavaca_Enero_comas.csv"
df = pd.read_csv(file_path, index_col=0, parse_dates=True)
df.index
```

```python
fig, ax = plt.subplots(figsize=(6, 3))

f1 = parse("2012-01-01")
f2 = f1 + pd.Timedelta("1d")

ax.plot(df.Ig,"r-", label="Ig")
ax.plot(df.Ib,"k--", label="Ib")
ax.plot(df.Id,"bo-", label="Id")

ax.legend()
ax.grid()
ax.set_xlim(f1, f2)

ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(mdates.AutoDateLocator()))
ax.set_ylabel("Irradiancia [$W/m^2$]")
ax.set_xlabel("Tiempo")
ax.set_title("Irradiancia en Cuernavaca")
fig.tight_layout()

fig.savefig("../../figs/irrad_cuernavaca.png", dpi=200)
fig.savefig("../../figs/irrad_cuernavaca.pdf", dpi=200)
```
