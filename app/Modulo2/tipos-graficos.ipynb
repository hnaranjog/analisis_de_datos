```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
file_path = "../../data/Cuernavaca_1dia_comas.csv"
df = pd.read_csv(file_path, index_col=0, parse_dates=True)
df.head()
```

```python
fig, ax = plt.subplots()

ax.plot(df.To)
ax.set_title("plot")
```

```python
# Gráfico de dispersión
# La función de dispersión (scatter) crea un gráfico donde cada punto representa un par de valores (x, y).
# En este caso, muestra cómo varía la temperatura exterior ('To') a lo largo del tiempo ('df.index').
# Es útil para visualizar la relación entre dos variables y detectar patrones, tendencias o valores atípicos.

fig, ax = plt.subplots()
ax.scatter(df.index, df.To)
ax.set_title("scatter")
```

```python
fig, ax = plt.subplots()

ax.bar(df.index, df.To)
ax.set_title("Bar")
```

```python
fig, ax = plt.subplots()

ax.stem(df.To)
ax.set_title("Stem")
```

```python
fig, ax = plt.subplots()

ax.fill_between(df.index, df.To.mean(), df.To, color="skyblue", alpha=0.4)
ax.plot(df.To, "r-")
ax.set_title("Fill Between")
```

```python
fig, ax = plt.subplots()

ax.stairs(df.To)
ax.set_title("Stairs")
```

```python
# Comparar df.To con df.Ws utilizando el grafico de dispersion, el valor de To azul y el valor de Ws rojo

fig, ax = plt.subplots()

ax.scatter(df.index, df.To, color="b", label="To")
ax.scatter(df.index, df.Ws, color="r", label="Ws")
ax.legend()
```
