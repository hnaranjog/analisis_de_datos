```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import glob as glob
```

```python
vels = glob.glob("../../data/obstacle/*")
vels.sort()
vels
```

```python
def transforma_imagen(imagen):
    img = Image.open(imagen)
    img = np.array(img)
    img = img[:,:,:3]
    return img



```
