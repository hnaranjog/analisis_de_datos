```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 0, 10)
y = np.linspace(-5, 5, 10)

X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots()
ax.scatter(X, Y)
```

```python
Z = X**2 + Y**2
Z.shape
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z)
```
