# La limpieza de Texte

1. Tirar TMP5
2. Tirar datos < 0
3. Renombrar columnas T1,,,T10
4. Resample promediado cada 10 minutos 


```python
file_path = "../../data/termopares/exterior.csv"
Text = pd.read_csv(file_path, index_col=0, parse_dates=True)
```

```python
nombres =  Text.columns.to_list()
nombres.remove("TMP5")
nombres
```

```python
Text = Text[nombres]
Text = Text[Text[nombres]>0]
```

```python
Text.columns = [ nombre.replace("MP", "") for nombre in nombres]
Text = Text.resample("10min").mean()
```

```python
Text.plot()
```

# Limpieza de Tcafe

1. Renombrar columnas
2. Resample promediado cada 10 minutos

# Ambos conjuntos con datos de 12 a 5 pm

```python
file_path = ("../../data/termopares/centro_cafeteria.csv")
Tcafe = pd.read_csv(file_path, index_col=0, parse_dates=True)
nombres = Tcafe.columns.to_list()
Tcafe = Tcafe[nombres]
Tcafe = Tcafe[Tcafe[nombres]>0]

Tcafe.columns = [nombre.replace("MP", "") for nombre in nombres]
Tcafe = Tcafe.resample("10min").mean()
Tcafe
```

```python
Tcafe.plot()
```
