```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "../../data/termopares/Tcafe-12a17h.csv"
Tcafe = pd.read_csv(file_path, index_col=0, parse_dates=True)

file_path = "../../data/termopares/Text-12a17h.csv"
Text = pd.read_csv(file_path, index_col=0, parse_dates=True)
```

```python
Ti = pd.concat([Tcafe, Text], axis=1, keys=["cafe", "ext"])
Ti["cafe"]
```

```python
Ti["cafe"].mean(axis=1)
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.legend()
ax.set_ylabel("Temperatura [$^oC$]")
ax.set_xlabel("Tiempo [mm-dd HH]")
ax.plot(Ti["cafe"].mean(axis=1),"r.", label="cafe")
ax.plot(Ti["ext"].mean(axis=1),"g.", label="ext")
```
