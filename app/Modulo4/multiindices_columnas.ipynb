```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "../../data/Ti_blanco.csv"
Ti_b = pd.read_csv(file_path, index_col=0, parse_dates=True)

file_path = "../../data/Ti_negro.csv"
Ti_n = pd.read_csv(file_path, index_col=0, parse_dates=True )
```

```python
Ti_b
```

```python
Ti_n
```

```python
pd.concat([Ti_b, Ti_n], axis=1)
```

```python
casos = pd.concat([Ti_b, Ti_n], axis=1, keys=["blanco", "negro"])
casos
```

```python
casos["blanco"]
```

```python
casos["negro"]
```

```python
fig, ax = plt.subplots()

ax.plot(casos["blanco"], c="k")
ax.plot(casos["negro"], c="r")
```

```python
fig, ax = plt.subplots()

ax.plot(casos["blanco"]["Ti_C"], c="k")
ax.plot(casos["negro"]["Ti_C"], c="r")
```

```python
from matplotlib import colors
```

```python
colores = casos.columns.levels[0]

fig, ax = plt.subplots(figsize=(10,3))

for color in colores:
    ax.plot(casos[color]["Ti_S"])
```

```python
m = pd.MultiIndex.from_product([["blanco"],Ti_b.index], names=["C","Fecha"])
```

```python
Ti_bm = Ti_b
Ti_bm.index = m
Ti_bm
```

```python
m = pd.MultiIndex.from_product([["negro"],Ti_n.index], names=["C","Fecha"])
```

```python
Ti_nm = Ti_n
Ti_nm.index = m
Ti_nm
```

```python
casos = pd.concat([Ti_bm, Ti_nm], axis=0)
casos
```
