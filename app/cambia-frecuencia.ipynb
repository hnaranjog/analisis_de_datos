```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Funcion resample: permite modificar la frecuencia de los datos

path_file = "../data/Temixco_2018_10Min.csv"
df = pd.read_csv(path_file, index_col=0, parse_dates=True)
df.info()
```

```python
# df.loc
# El método loc de pandas permite seleccionar datos de un DataFrame utilizando etiquetas de los índices y nombres de columnas.
# Por ejemplo, df.loc["2018-03-10", "To"] selecciona todos los valores de la columna "To" para la fecha "2018-03-10".

# plot()
# La función plot de pandas se utiliza para visualizar datos en forma de gráficos. 
# Permite personalizar el tipo de gráfico, el tamaño, el estilo y otros parámetros.
# En este caso, se grafica la serie temporal de la columna "To" para el día seleccionado con un estilo de puntos.

df.loc["2018-03-10", "To"].plot(figsize=(10,2), style="o")
```

```python
df.resample("h").max()
```

```python
# Resample: permite modificar la frecuencia de los datos 
# closed: indica si el intervalo de tiempo está cerrado por la izquierda o por la derecha
# Por ejemplo, en resample("h", closed="right"), cada intervalo horario incluye el extremo derecho (final del intervalo) pero no el izquierdo.
# Si usas closed="left", el intervalo incluirá el extremo izquierdo (inicio del intervalo) pero no el derecho.
# Esto afecta qué filas se consideran dentro de cada grupo al hacer el resample.
# max: devuelve el valor máximo de cada intervalo
df.resample("h", closed="right").max()
```

```python
# Gráficos de la serie temporal
# 1. La primera gráfica muestra los valores originales de temperatura exterior ("To") cada 10 minutos para el 10 de marzo de 2018.
# 2. La segunda gráfica (roja) muestra el valor máximo de "To" por cada hora de ese día.
# 3. La tercera gráfica (verde) muestra el valor promedio de "To" por cada hora de ese día.
df.loc["2018-03-10", "To"].plot(figsize=(10,2), style=".")
df.loc["2018-03-10", "To"].resample("h").max().plot(figsize=(10,2), style="o", color="red")
df.loc["2018-03-10", "To"].resample("h").mean().plot(figsize=(10,2), style="o", color="green")
```

```python
# Resample: permite modificar la frecuencia de los datos
df.To.resample("D").mean()
```
